cambios en las filas del dataframe y reindexar la enumeración acorde. Usar el dataframe
original (sin promedios).


```python
def agregar_alumno(registro, alumno, legajo, nota1, nota2):
  nuevo_alumno = {
      'Alumnos' :  alumno        ,
      'Legajo'  :  legajo        ,
      'Nota 1'  :  nota1         ,
      'Nota 2'  :  nota2         ,
      'promedio':  (nota1 + nota2)/2             
  }

  return registro.append(nuevo_alumno, ignore_index=True)
```
