### **a) Gráfico para visualizar las ganancias según el mes.**</font>

### <font color="#17202A">Datos para precisar el gráfico</font>

```python
# Importar el módulo pandas con el alias pd
import pandas as pd
# Declarar la ruta
ruta = '/content/'
# Declarar el archivo
archivo = 'company_sales_data.csv'
# Crear el DataFrame
df = pd.read_csv(ruta+archivo)
# Mostrar el dataFrame
df.head()
```

```python
# Definir DataFrame de meses y ganancias
month_profit = pd.DataFrame(df[['month_number','total_profit']])
# Mostrar DataFrame
month_profit.head()
```

```python
# Ordenar los valores en 'total_profit'
month_profit.sort_values(by='total_profit', ascending=False, ignore_index=True, inplace=True)
# Mostrar DataFrame con el mes de mayor ganancia
month_profit.head()
```

```python
# Variables para graficar
meses = pd.DataFrame(month_profit['month_number'])
ganancias = pd.DataFrame(month_profit['total_profit'])
ganancia_maxima = pd.DataFrame(month_profit['total_profit'],[0])
ganancia_maxima.head()
```

### <font color='#17202A'>Gráfico</font>

```python
# Importar el módulo pyplot con el alias plt
import matplotlib.pyplot as plt
# Crear la figura y los ejes
fig, ax = plt.subplots(figsize = (20, 6))
# Dibujar puntos
ax.scatter(x = [ganancias], y = [meses])
# Mostrar cada numero de los meses
plt.yticks(range(0, 13, 1))
# Añadir Formato al gráfico
# Titulo
fig.suptitle('a) Graficar las ganancias según el mes e indicar cuál fue el mes de mayor ganancia con un punto.',fontsize=16)
# Eje X
ax.set_xlabel('Ganancias')
# Eje Y
ax.set_ylabel('Meses')
# Grilla / cuadricula
ax.grid(True)
# Marcar mes de mayor ganancia
plt.plot(ganancia_maxima,11, marker="o", color="gold", label = 'Mes de mayor ganancia')
# Conectar puntos con linea
plt.plot(ganancias,meses, linestyle='solid', label = 'Resto de los meses y ganancias')
# Leyenda
ax.legend(loc='upper left')
# Guardar el gráfico en formato png
plt.savefig('mes-mayorGanancia.png')
# Mostrar el gráfico
plt.show()
```

<font color="#424949">2) Para el dataset del ejercicio 1, sumar todas las ventas para cada producto durante el año. Luego hacer un gráfico de torta mostrando el porcentaje de ventas de cada categoría respecto al total.</font>

### <font color="#17202A">Datos para precisar el gráfico</font>

```python
datos.head()
```

### <font color='#17202A'>Gráfico</font>

```python
# Importar modulos
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib import colors

# Sumar ventas de cada producto
facecream   = sum(datos['facecream'])
facewash    = sum(datos['facewash'])
toothpaste  = sum(datos['toothpaste'])
bathingsoap = sum(datos['bathingsoap'])
shampoo     = sum(datos['shampoo'])
moisturizer = sum(datos['moisturizer'])
# Sumar total
total       = facecream+facewash+toothpaste+bathingsoap+shampoo+moisturizer

# Agrupar en listas para poder graficar y nombrar cada porcentaje
nombres     = ['facecream','facewash','toothpaste','bathingsoap','shampoo','moisturizer']
productos   = [facecream,facewash,toothpaste,bathingsoap,shampoo,moisturizer]

# Desfasaje en el producto 'bathingsoap'
desfase = (0, 0, 0, 0.1, 0, 0)

# Estilos del pastel
normdata = colors.Normalize(min(productos), max(productos))
colormap = cm.get_cmap("Blues")
colores =colormap(normdata(productos))

# Ajustes finales para las etiquetas, colores desfasaje, etc
plt.pie(productos, labels=nombres, autopct="%0.1f %%", explode=desfase, colors=colores)
# Ajustar los ejes
plt.axis("equal")
# Tamaño
fig = plt.gcf()
fig.set_size_inches(8,8)
# Mostrar
plt.show()
# Referncias de las sumas
print(f'''
Facecream   : {facecream}
Facewash    : {facewash}
Toothpaste  : {toothpaste}
Bathingsoap : {bathingsoap}
Shampoo     : {shampoo}
Moisturizer : {moisturizer}
Total       : {total} ''')
```
